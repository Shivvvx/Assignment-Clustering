**Q1.What is unsupervised learning in the context of machine learning ?**
 - Unsupervised learning is a type of machine learning in which the model is trained on data without any labeled outputs. Instead of learning from predefined answers, the algorithm tries to identify hidden patterns, structures, or relationships within the data on its own. It clusters similar data points together, reduces dimensionality to simplify complex datasets, and detects unusual patterns or anomalies. Common techniques include K-Means clustering, Hierarchical clustering, PCA, and anomaly detection methods. Unsupervised learning is especially useful when you want to explore data, segment customers, group similar products, or understand underlying trends without prior knowledge of the correct output.

**Q2. How does K-Means clustering algorithm work?**
 - K-Means clustering is an unsupervised learning algorithm that groups data into **K distinct clusters** based on similarity. It works by first randomly selecting **K points as initial centroids**. Then, in an iterative process, each data point is assigned to the nearest centroid, forming K clusters. Next, the algorithm recalculates the centroids by taking the **mean position of all points** in each cluster. These two steps—assignment and centroid update—repeat until the centroids stop changing significantly or a maximum number of iterations is reached. The goal of K-Means is to minimize the **within-cluster variance**, meaning each cluster becomes as compact and distinct as possible.


**Q3.Explain the concept of a dendrogram in hierarchical clustering ?**
 - A **dendrogram** is a tree-like diagram used in hierarchical clustering to visually represent how clusters are formed step-by-step. It starts by treating each data point as its own cluster and then progressively merges the closest clusters based on a distance measure. The vertical lines in the dendrogram show where clusters combine, and the height of each merge represents the **distance or dissimilarity** between the clusters being joined. By looking at the dendrogram, you can see the entire clustering process, understand how similar different groups are, and decide the optimal number of clusters by “cutting” the tree at a chosen height.


**Q4.What is the main difference between K-Means and Hierarchical Clustering?**
 - The main difference between **K-Means** and **Hierarchical Clustering** is how they form clusters. **K-Means** is a **partitioning method** that requires you to specify the number of clusters (K) in advance and then groups data by repeatedly assigning points to the nearest centroid. In contrast, **Hierarchical Clustering** builds a **tree-like structure of clusters** without needing to predefine the number of clusters; you can choose the number later by cutting the dendrogram. K-Means is faster and better for large datasets, while hierarchical methods are more interpretable but computationally expensive for large data.


**Q5. What are the advantages of DBSCAN over K-Means ?**
 - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages over K-Means:

1. **Does not require specifying the number of clusters** beforehand, unlike K-Means which needs K in advance.
2. **Can find clusters of arbitrary shapes**, while K-Means is limited to spherical or convex clusters.
3. **Handles noise and outliers effectively** by labeling them as “noise,” whereas K-Means forces every point into a cluster.
4. **Works well for unevenly sized and differently shaped clusters**, whereas K-Means struggles when clusters have varying density or size.

Overall, DBSCAN is more flexible and robust in real-world datasets where cluster shapes and densities vary.


**Q6. When would you use Silhouette Score in clustering?**
 - You would use the **Silhouette Score** in clustering to **evaluate how well your data points have been grouped** and to help determine the **optimal number of clusters**. The score measures how similar a point is to its own cluster compared to other clusters. It ranges from **-1 to +1**, where a higher value means better-defined clusters. A high silhouette score indicates that clusters are **well separated and cohesive**, while a low or negative score suggests that data points are misclassified or clusters overlap. It is commonly used when tuning K in K-Means, choosing the best clustering model, or comparing different clustering algorithms.


**Q7. What are the limitations of Hierarchical Clustering?**
 - Hierarchical clustering has several limitations:

1. **High computational cost** — It is slow and memory-intensive for large datasets because it requires calculating and storing all pairwise distances.
2. **No ability to correct early mistakes** — Once clusters are merged or split, the decision cannot be undone, which can lead to suboptimal results.
3. **Sensitive to noise and outliers** — A few noisy points can distort the entire clustering structure.
4. **Difficult to scale** — It works well for small datasets but becomes impractical for large-scale real-world data.
5. **Choosing the optimal number of clusters is subjective** — It depends on how you “cut” the dendrogram, which can vary based on interpretation.



**Q8. Why is feature scaling important in clustering algorithms like K-Means?**
 - Feature scaling is important in clustering algorithms like **K-Means** because the algorithm relies on **distance calculations (usually Euclidean distance)** to group data points. If features are on different scales—for example, age ranging from 20–60 and income ranging from 20,000–2,00,000—the feature with the larger numeric range will dominate the distance measure. This causes the clustering results to become biased toward features with bigger values. Scaling methods like **Standardization (Z-score)** or **Min-Max Normalization** ensure all features contribute equally, leading to more accurate and meaningful clusters.


**Q9. How does DBSCAN identify noise points?**
 - DBSCAN identifies noise points based on **density criteria**. It looks for regions in the data where points are closely packed together. A point is classified as **noise (or an outlier)** if it does *not* satisfy the following conditions:

1. **Core Point Condition:**
   The point must have at least `min_samples` points (including itself) within a distance of `eps`.
   If it doesn’t meet this density requirement, it cannot be a core point.

2. **Border Point Condition:**
   If a point is not a core point but lies within the `eps` radius of a core point, it becomes a **border point**, not noise.




**Q10. Define inertia in the context of K-Means?**
 - Inertia in the context of **K-Means** is a measure of how well the algorithm has clustered the data. It represents the **sum of squared distances between each data point and its assigned cluster centroid**. Lower inertia means that data points are closer to their respective centroids, indicating **tighter, more compact clusters**. During model selection, we try different values of K and look at inertia to help decide the optimal number of clusters—often using the **Elbow Method**, where the rate of inertia decrease slows down, suggesting a good choice for K.


**Q11. What is the elbow method in K-Means clustering?**
 - The **Elbow Method** in K-Means clustering is a technique used to determine the **optimal number of clusters (K)**. It works by running K-Means for different values of K and plotting **Inertia (within-cluster sum of squared distances)** against K. As K increases, inertia always decreases, but after a certain point, the improvement becomes small. The plot forms a shape like an **elbow**, and the point where the curve starts to flatten is considered the optimal K. This is because adding more clusters beyond this point does not significantly improve clustering quality, making the elbow point a good balance between simplicity and accuracy.

**Q12.Describe the concept of "density" in DBSCAN?**
 - In DBSCAN, **density** refers to how tightly data points are packed together within a region. The algorithm defines density using two important parameters:

1. **eps (ε):**
   The maximum radius around a point to search for neighboring points.

2. **min_samples:**
   The minimum number of points required within the eps radius for that region to be considered dense.

A region is considered **dense** if a point has at least `min_samples` points (including itself) within its eps neighborhood. Such points are called **core points**. Clusters form by connecting core points that lie within each other's eps distance. Points in sparse areas that don’t meet the density requirement are treated as **noise or outliers**.



**Q13.Can hierarchical clustering be used on categorical data?**
 - Yes, **hierarchical clustering can be used on categorical data**, but not directly with standard distance measures like Euclidean distance. For categorical variables, you need to use **appropriate similarity or distance metrics**, such as:

* **Hamming distance**
* **Jaccard distance**
* **Simple matching coefficient (SMC)**

These metrics measure how similar or different categorical values are. Once you compute a proper distance matrix using these methods, hierarchical clustering can be applied in the usual way.




**Q14.What does a negative Silhouette Score indicate?**
 - A **negative Silhouette Score** indicates that a data point is likely **misclassified** — meaning it is closer, on average, to points in a **different cluster** than to points in its own cluster. This suggests poor clustering quality. Negative scores usually happen when:

* Clusters overlap significantly
* Wrong number of clusters (K) is chosen
* Data points are assigned to the wrong cluster
* The structure of the data is not suitable for the chosen algorithm



**Q15.Explain the term "linkage criteria" in hierarchical clustering**
 - In hierarchical clustering, **linkage criteria** refers to the method used to determine the **distance between two clusters** when deciding which clusters to merge next. Since clusters contain multiple points, we need a rule to calculate the cluster-to-cluster distance. Common linkage criteria include:

1. **Single Linkage:**
   Distance between the **closest** pair of points from the two clusters.

2. **Complete Linkage:**
   Distance between the **farthest** pair of points in the clusters.

3. **Average Linkage:**
   Distance between clusters is the **average** of all pairwise distances.

4. **Ward’s Method:**
   Merges clusters by minimizing the **increase in total within-cluster variance**.


**Q16.Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?**
 - K-Means tends to struggle when clusters come in **different sizes, shapes, or densities** because it relies on a simple assumption: clusters are roughly **spherical, equally sized, and evenly dense**. When the data breaks these assumptions, a few things can go wrong:

* **Larger clusters dominate the centroid movement**, pulling centroids away from smaller clusters.
* **Dense clusters get mixed with sparse ones**, because K-Means treats all distances uniformly.
* **Irregular or non-spherical shapes** (like elongated or crescent-shaped clusters) can't be captured well, since the algorithm forces clusters to form around a mean point.
* **Overlapping clusters** confuse assignments because K-Means always assigns each point to the nearest centroid, even if the true structure is more complex.



**Q17.What are the core parameters in DBSCAN, and how do they influence clustering?**

 - **Spherical (round)**

 - **Equal in density**

Since K-Means uses **Euclidean distance** and focuses on minimizing the distance to the centroid, it struggles when:

 - Some clusters are **dense** and others are **sparse**

 - Some clusters are **large** and others are **small**

 - Clusters have **irregular or elongated shapes**

 - Outliers pull centroids away from the correct position

**Q19.What is agglomerative clustering?**
 - **Agglomerative clustering** is a type of **hierarchical clustering** that builds clusters **from the bottom up**. It starts by treating each data point as its **own individual cluster**. Then, step-by-step, it **merges the two closest clusters** based on a chosen distance metric and linkage method (like single, complete, or average linkage). This merging continues until all points form one large cluster or until a desired number of clusters is reached. The entire merging process can be visualized using a **dendrogram**, which shows how clusters combine at each step.

```python
# Q21.Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot?
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 centers
X, y = make_blobs(n_samples=500, centers=4, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means Clustering on Synthetic Data")
plt.show()
```

```python
# Q29.Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart?
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, y = make_blobs(n_samples=500, centers=4, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
ks = [2, 3, 4, 5]
scores = []

for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Bar chart
plt.figure(figsize=(8, 6))
plt.bar(ks, scores)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for K-Means (k = 2 to 5)")
plt.show()
```

```python
# Q31. import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np

# Generate overlapping synthetic data
X, y = make_blobs(n_samples=400, centers=3, cluster_std=2.0, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Create meshgrid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))

# Predict cluster for each grid point
grid_points = np.c_[xx.ravel(), yy.ravel()]
Z = kmeans.predict(grid_points)
Z = Z.reshape(xx.shape)

# Plot decision boundaries
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3)

# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200)

plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Q32.Load the Digits dataset and apply DBSCAN after reducing dimensions with t-SNE. Visualize the results?
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np

# Load digits dataset
digits = load_digits()
X = digits.data

# Standardize features before t-SNE (good practice)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce dimensions to 2D using t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42, init='pca', learning_rate='auto')
X_2d = tsne.fit_transform(X_scaled)

# Apply DBSCAN on the 2D embedding
# eps chosen conservatively for t-SNE scale; min_samples set to 5
dbscan = DBSCAN(eps=2.5, min_samples=5)
labels = dbscan.fit_predict(X_2d)

# Prepare plot
plt.figure(figsize=(9, 6))

# Identify noise points
noise_mask = labels == -1
core_mask = labels != -1

# Scatter cluster points colored by label
plt.scatter(X_2d[core_mask, 0], X_2d[core_mask, 1], c=labels[core_mask], s=30)

# Overlay noise points with 'x' marker
if noise_mask.any():
    plt.scatter(X_2d[noise_mask, 0], X_2d[noise_mask, 1], marker='x', s=60)

plt.title("DBSCAN on t-SNE reduced Digits dataset")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")

# Print cluster counts to console (excluding noise)
unique, counts = np.unique(labels, return_counts=True)
cluster_counts = dict(zip(unique, counts))
print("Cluster sizes (label: count):", cluster_counts)

plt.show()
```

```python
# Q34.LOad the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot?
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load dataset
data = load_breast_cancer()
X = data.data

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Compare inertia for K = 2 to 6
Ks = range(2, 7)
inertias = []

for k in Ks:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Line plot
plt.figure(figsize=(8, 6))
plt.plot(Ks, inertias, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Inertia Values for K-Means (Breast Cancer Dataset)")
plt.grid(True)
plt.show()
```

```python
# Q40. Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D scatter plot.
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load Digits dataset
digits = load_digits()
X = digits.data

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce to 3 principal components
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# Apply KMeans
kmeans = KMeans(n_clusters=10, random_state=42)
labels = kmeans.fit_predict(X_pca)

# 3D Scatter Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels)

ax.set_title("3D PCA Scatter Plot with KMeans Clusters (Digits Dataset)")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")

plt.show()
```

```python
# Q42.oad the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering. Visualize in 2D
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load dataset
data = load_breast_cancer()
X = data.data

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce to 2D using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels = agg.fit_predict(X_pca)

# Plot 2D results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
plt.title("Agglomerative Clustering on Breast Cancer Data (2D PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Q46.Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load Wine dataset
data = load_wine()
X = pd.DataFrame(data.data, columns=data.feature_names)

# Use only first 4 features
X_subset = X.iloc[:, :4]

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_subset)

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Add cluster labels to dataframe
X_subset["cluster"] = labels

# Create seaborn pairplot
sns.pairplot(X_subset, hue="cluster", diag_kind="kde")
plt.show()
```

```python
# Q47.Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate noisy blobs
X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42)

# Scale for DBSCAN
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
db = DBSCAN(eps=0.5, min_samples=5)
labels = db.fit_predict(X_scaled)

# Count clusters (excluding noise)
unique = set(labels)
if -1 in unique:
    unique.remove(-1)

cluster_count = len(unique)
noise_count = list(labels).count(-1)

cluster_count, noise_count
```

```python
# Q48.Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

# Load Digits dataset
digits = load_digits()
X = digits.data

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, perplexity=30, random_state=42, init='pca', learning_rate='auto')
X_2d = tsne.fit_transform(X_scaled)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_2d)

# Plot clusters
plt.figure(figsize=(9, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels)
plt.title("Agglomerative Clustering on t-SNE Reduced Digits Dataset")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
